## SelectHubbard



 - If `control` is off, the circuit has no effect. 
 - If `U, V == 1, 0` then `(p, alpha) == (q, beta)` and we apply $-Z_{p,\alpha}$ to `target`.
 - If `U, V == 0, 1` then `p == q` but `alpha == 0 and beta == 1`, so we apply $Z_{p,0} Z_{q,1}$.
 - If `U, V == 0, 0` then we apply $-X_{p,\alpha} Z X_{q,\alpha}$ or $-Y_{p,\alpha} Z Y_{q,\alpha}$ for $p<q$ and $p>q$, resp.

```python
from cirq_qubitization.hamiltonians.hubbard_qubitization import SelectHubbard

g = cq_testing.GateSystem(
    SelectHubbard(4, 1)
)

display_gate_and_compilation(g)
```
